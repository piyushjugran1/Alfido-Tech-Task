```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('superstore_final_dataset (1).csv', encoding='latin1')
df.head()
```

```python
print(df.isnull().sum())

df.drop_duplicates(inplace=True)

df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, format='mixed')
```

```python
print(df.columns)

total_revenue = df['Sales'].sum()

aov = df['Sales'].mean()

# The 'Profit' column does not exist in the DataFrame.
# total_profit = df['Profit'].sum()
# profit_margin = (total_profit / total_revenue) * 100

print(f"Total Revenue: {total_revenue}")
print(f"Average Order Value: {aov}")
# print(f"Profit Margin: {profit_margin:.2f}%") # Removed as 'Profit' column is missing
```

```python
plt.figure(figsize=(12,6))
df.set_index('Order_Date').resample('ME')['Sales'].sum().plot()
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Total Sales by Region')
plt.show()
```

```python
top_products = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Products by Sales')
plt.ylabel('Sales')
plt.show()
```
